```python
MY_STATES = ['Colorado', 'Oklahoma', 'Utah', 'California', 'Nevada', "New_Mexico", "Arizona", "North_California"]
PINNACLE_URLS = "/home/vince/Documents/SmartFires/osm_fitness/Azira/get-urls/urls.txt"
```

```python
def get_geojson_files(MY_STATES):
    geojson_files = []
    base_dir = Path('/home/vince/Documents/SmartFires/osm_fitness')
    folders = [p for p in base_dir.iterdir() if p.is_dir() and (sum([s in p.name for s in MY_STATES]) > 0)]

    for folder in folders:  
        curr_files = list(folder.glob('*.geojson'))
        for json_path in curr_files:
            geojson_files.append(json_path)  
    
    return [(p.stem.replace(' ', '_'), p) for p in geojson_files]    
```

```python
def get_not_downloaded(urls_file, states):
    geojson_files = get_geojson_files(states)
    with Path(urls_file).open() as f:
        urls =f.read()

    not_downloaded = [f for f in geojson_files if f[0] not in urls]
    return not_downloaded
```

```python
not_download_list = get_not_downloaded(PINNACLE_URLS, MY_STATES)[:-1] # skip st george
not_download_list
```

```python
with open("not_downloaded_1224.txt", "w") as f:
    for name, path in not_download_list: 
        f.write(f"{path}\n")
```
