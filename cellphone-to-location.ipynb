```python
import folium
import json
import os
import geopandas as gpd
file_path = "Idaho/Twin Falls.geojson"
path = "unified_geopackages_no_missing/state_16_unified.gpkg"

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Grab cities from geojson
cities = [
    c.strip()
    for c in data["features"][0]["properties"]["all_cities"].split(",")
]
```

```python
out_dir = "unified_geopackages_no_missing/split_geopackages"
os.makedirs(out_dir, exist_ok=True)

gdf = gpd.read_file(path, layer="unified_park_area") 

for city in cities:
    city_gdf = gdf[gdf["city_name"] == city].copy()

    if city_gdf.empty:
        continue

    # Optional: normalize CRS for export
    city_gdf = city_gdf.to_crs(4326)

    out_path = os.path.join(out_dir, f"{city}.gpkg")

    city_gdf.to_file(
        out_path,
        layer=city.replace(" ", "_").lower(),
        driver="GPKG"
    )

    print(f"Saved {city}: {len(city_gdf)} features")
```
